```python
import requests
from bs4 import BeautifulSoup

url = "http://164.100.60.165/comm/dsa.html"

# Send a GET request to the URL
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find all <option> tags within the document
options = soup.find_all("option")

# Extract the values and print them
tags=[]
for option in options:
    option_value = option["value"]
    print(option_value)
    tags.append(option_value)
```

```python
base_url = "http://164.100.60.165/comm/2021-22/dsa/"

url_list = ["{}{}{}".format(base_url, str(tags[i]),"/dsa.pdf") for i in range(0,len(tags))]
s=[]
for url in url_list:
    print (url)
    s.append(url)
```

```python
import requests
import os

def download_files_from_urls(urls, folder_path):
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    for url in urls:
        # Extract the filename from the URL
        filename = url.split("/")[-2]

        # Set the file path to save the downloaded file
        file_path = os.path.join(folder_path, filename)

        # Send a GET request to download the file
        response = requests.get(url)

        # Save the file to the specified path
        with open(file_path, "wb") as file:
            file.write(response.content)

        print(f"File downloaded: {filename}")



folder_path = "DSA data reports"

download_files_from_urls(s, folder_path)
```

```python
import requests
import os
import urllib.parse

def download_files_from_urls(urls, folder_path):
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    for url in urls:
        # Extract the filename from the URL
        filename = url.split("/")[-2]

        # Add .pdf extension if missing
        if not filename.lower().endswith(".pdf"):
            filename += ".pdf"

        # Set the file path to save the downloaded file
        file_path = os.path.join(folder_path, filename)

        # Send a GET request to download the file
        response = requests.get(url)

        # Save the file to the specified path
        with open(file_path, "wb") as file:
            file.write(response.content)

        print(f"File downloaded: {filename}")



folder_path = "DSA data pdf reports"

download_files_from_urls(s, folder_path)
```

```python
import os
import openpyxl
from PyPDF2 import PdfReader

def extract_pdf_data_to_excel(pdf_file, output_file):
    # Read the PDF file and extract the data
    reader = PdfReader(pdf_file)

    # Create a new Excel workbook
    workbook = openpyxl.Workbook()

    # Create a new sheet in the workbook
    sheet = workbook.active

    # Write the data to the sheet
    for index, page in enumerate(reader.pages, start=1):
        sheet.title = f"Page {index}"
        text = page.extract_text()
        rows = text.strip().split("\n")
        for row in rows:
            sheet.append(row.split())

    # Save the workbook to the specified output file
    workbook.save(output_file)

# Folder paths
folder_path = "DSA data pdf reports"
output_folder = "DSA data excel format"

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Get the list of PDF files in the folder
pdf_files = [file for file in os.listdir(folder_path) if file.endswith(".pdf")]

# Process each PDF file
for pdf_file in pdf_files:
    # Construct the paths for input and output files
    pdf_path = os.path.join(folder_path, pdf_file)
    output_path = os.path.join(output_folder, pdf_file.replace(".pdf", ".xlsx"))

    # Extract data from PDF and save to Excel
    extract_pdf_data_to_excel(pdf_path, output_path)

    print(f"Processed: {pdf_file}")
```
