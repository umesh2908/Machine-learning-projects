```python
##Data Scrapper function which can be called to scrape data 

def dataScrapper(date):
    base_url = "https://www.delhisldc.org/Loaddata.aspx?mode="

    # Add range (0,number of day) to iterate the date from page list.
    url_list = ["{}{}".format(base_url, str(date[i])) for i in range(0,len(date))]
    s=[]
    for url in url_list:
        print (url)
        s.append(url)

    Month_2021=[]    ## list of daily tables
    for i in s:
        data=pd.read_html(i,header=0)
        table=data[2]
        table['date']=i[-10:]
        Month_2021.append(table)

    dataset=pd.concat(Month_2021)       ## dataframe of all daily tables
    
    return dataset
```
